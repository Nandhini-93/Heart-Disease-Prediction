Machine Learning-Random Forest classifier

```python
# Creating features and labels-->(X,y)
X=heart_disease.drop("target",axis=1) # features   axis=1-->vertical and column, axis=0-->horizontal and row
y=heart_disease["target"]  # labels
```

```python
# choosing the right model---> RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100)
clf.get_params()
```

```python
# Fit the model to training data and split into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
clf.fit(X_train,y_train)
X_train
```

```python
# Make prediction
y_pred=clf.predict(X_test)
y_pred
```

```python
# Check accuracy
clf.score(X_train,y_train) #1.0---> 100% accuracy
```

```python
clf.score(X_test,y_test)
```

```python
# Metrics
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print(classification_report(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix is:\n",cm)
```

```python
# improving the model
np.random.seed(42)
for i in range(10,100,10):
  print(f"Trying model with {i}estimators")
  clf=RandomForestClassifier(n_estimators=i).fit(X_train,y_train)
  print(f"Model accuracy on the test set:{clf.score(X_test,y_test)*100}%")
```

plotting the confusion matrix

```python
conf_matrix = pd.DataFrame(data = cm,
                           columns = ['Predicted:0', 'Predicted:1'],
                           index =['Actual:0', 'Actual:1'])
plt.figure(figsize = (10, 6))
sns.heatmap(conf_matrix, annot = True, fmt = 'd', cmap = "Greens", linecolor="Blue", linewidths=1.5)
plt.show()

```
